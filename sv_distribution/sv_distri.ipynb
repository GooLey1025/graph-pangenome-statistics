```r
# 加载必要库
library(ggplot2)

# 读取数据
data <- read.table("37maize_sv_distribution.tsv", header = FALSE, sep = "\t")
colnames(data) <- c("Length", "Count")  # 为列命名

# 筛选 Length > 50 的数据
data <- data[data$Length > 50, ]

# 计算加权平均值和中位数
average_length <- round(weighted.mean(data$Length, data$Count))  # 加权平均并四舍五入
median_length <- round(data$Length[which(cumsum(data$Count) >= sum(data$Count) / 2)][1])  # 加权中位数并四舍五入

# 打印平均值和中位数
cat("Average Length:", average_length, "bp\n")
cat("Median Length:", median_length, "bp\n")

# 绘制面积图
ggplot(data, aes(x = Length, y = Count)) +
  geom_area(fill = "#46b49a", alpha = 0.7) +  
  scale_x_log10(expand =c(0.01,0)) +  # 对数横轴（适用于 Length 的跨度很大）
  scale_y_continuous(expand=c(0.01,0))+
  labs(
    x = "Length of SVs",  # 横坐标标题
    y = "Number of SVs"  # 纵坐标标题
  ) +
  # 添加红色虚线
  geom_vline(xintercept = median_length, color = "red", linetype = "dashed") +
  geom_vline(xintercept = average_length, color = "#002fff", linetype = "dashed") +
  # 添加水平注释
  annotate(
    "text", x = median_length, y = max(data$Count) * 0.9, 
    label = paste("Median:", median_length, "bp"), 
    color = "red", angle = 0, hjust = 0.5,size=5
  ) +
  annotate(
    "text", x = average_length, y = max(data$Count) * 0.8, 
    label = paste("Average:", average_length, "bp"), 
    color = "#002fff", angle = 0, hjust = -0.1,size=5
  ) +
  theme_bw() +  # 使用带框的经典主题
  theme(
    axis.title.x = element_text(size = 18),  # 横坐标标题字体大小
    axis.text.x = element_text(size = 16),  # 横坐标刻度字体大小
    axis.text.y = element_text(size = 16),  # 纵坐标刻度字体大小
    axis.title.y = element_text(size = 18)  # 纵坐标标题字体大小
  )
```
